# program to find the euclidean distance between two coordinates.Take both the coordinates from the user as input.

```python
# in this code i am importing python module to perform the square root 

import math
x1 = int(input("enter the x1"))
x2 = int(input("enter the x2"))
y1 = int(input("enter the y1"))
y2 = int(input("enter the y2"))

dis = math.sqrt((x2-x1)**2 + (y2-y1)**2)
print(dis)
```

# program to find the simple interest when the value of principle,rate of interest and time period is provided by the user.

# program that will tell the number of dogs and chicken are there when the user will provide the value of total heads and legs.
## For example: Input: heads -> 4 legs -> 12
## Output: dogs -> 2 chicken -> 2

```python
# here first we made equation for both total number of heads and total number of legs 
# after this take number of heads and number of legs as input
h = int(input("enter the number of heads"))
l = int(input("enter the number of legs"))
d = (l-2*h)/2
c = (4*h-l)/2
print("number of dogs",d)
print("number of chicken",c)
```

# program to find the sum of squares of first n natural numbers where n will be provided by the user.

```python
# here we initialize sum from zero and run a loop from starting of natural to the number input from user 
# one by one we add the square of that number

n = int(input("enter the number"))
sum = 0
for i in range(1,n+1):
    sum += i**2
print(sum)
```

# Given the first 2 terms of an Arithmetic Series.Find the Nth term of the series. Assume all inputs are provided by the user.

```python
# here we have first two term of ap 
# by using nth term formula we find the nth term of ap 
# we take first term,second term and nth term as input

a1 = int(input("enter the first term"))
a2 = int(input("enter the second term"))
n = int(input("enter the nth term you want"))
d = a2 - a1

an = a1 + ((n-1)*d)
print("print the nth term of ap",an)

```

#  Given 2 fractions, find the sum of those 2 fractions.Take the numerator and denominator values of the fractions from the user.

```python
# here we take numerator and denominator as input then perform addition on it
num1 = int(input("enter the num of first fraction"))
num2 = int(input("enter the num of second fraction"))
deno1 = int(input("enter the deno of first fraction"))
deno2 = int(input("enter the deno of second fraction"))
sum = (num1/deno1) + (num2/deno2)
print("sum of two fraction",sum)
```
