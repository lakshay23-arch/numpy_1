# Take a user input as integer N. Find out the sum from 1 to N. If any number if divisible by 5, then skip that number. And if the sum is greater than 300, don't need to calculate the sum further more. Print the final result. And don't use for loop to solve this problem.

```python
n = int(input("enter the number"))
i = 1
sum = 0
while i != n+1:
    if i % 5 == 0:
        i = i+1
        continue
    
    else:
        sum = sum + i
        if sum > 300:
            sum = sum -i
            break
        i = i+ 1

print(sum)
```

# program that keeps on accepting a number from the user until the user enters Zero. Display the sum and average of all the numbers.

```python
n = int(input("enter the number"))
count = 0
sum = 0
while n != 0:
    sum = sum + n
    count = count + 1
    n = int(input("enter the number"))


print("sum of all number",sum)
ave = sum/count
print("average of all number",ave)
```

# to check given string is pallendrome or not

```python
st = input("enter the string: ")
l = int(input("enter the length of string: "))

# Reverse the string
a = st[-1::-1]

for i in range(0, l):
    if st[i] != a[i]:  # 1. Check if they are DIFFERENT
        print("it is not a palindrome")
        break  # 2. Stop immediately and SKIP the else block
else:
    # 3. This only runs if the loop finished completely WITHOUT hitting the break
    print("It IS a palindrome!")
```
