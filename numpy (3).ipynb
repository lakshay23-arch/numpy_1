# how numpy array is faster then the python list

```python
# this is python list
a = [ i for i in range(10000000)]
b = [ i for i in range(10000000,20000000)]

c = []
import time 
start = time.time()
for i in range(10000000):
    c.append(a[i] + b[i])
print(time.time() - start)
```

```python
# now numpy array
import numpy as np
a = np.arange(10000000)

b = np.arange(10000000,20000000)

start = time.time()
c = a+b
print(time.time() - start)


```

### like this numpy array takes less time to to append element in list

# Boolean indexing

```python
axp = np.random.randint(1,100,24).reshape(6,4)
axp
```

```python
# number divisible by 2
axp[axp%2 == 0]
```

```python
# number greater then 50 and divisible by 2
axp[(axp%2 == 0) & (axp>50)]
```

```python
# find the number which is divisible by 7
axp[axp % 7 == 0]
```
